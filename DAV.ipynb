```python
import numpy as np
import pandas as pd
```

```python

amount_array = np.array(df['Amount'])
boxes_array = np.array(df['Boxes Shipped'])
```

```python
print("Total Revenue:", np.sum(amount_array))
print("Average Revenue:", np.mean(amount_array))
print("Standard Deviation:", np.std(amount_array))
print("Minimum Revenue:", np.min(amount_array))
print("Maximum Revenue:", np.max(amount_array))
print("Median Revenue:", np.median(amount_array))
```

```python
# Revenue > 1000
high_sales = df[np.array(df['Amount']) > 1000]
print(high_sales)
```

```python
# Revenue per box using NumPy
revenue_per_box = amount_array / boxes_array
df['Revenue_per_Box_np'] = revenue_per_box
```

```python
# Reshape revenue array to 2D
reshaped = amount_array[:10].reshape(2, 5)
print("Reshaped Array (2x5):\n", reshaped)
```

```python
# Mask: Which entries are above average?
above_avg_mask = amount_array > np.mean(amount_array)
print("Above Average Sales:\n", df[above_avg_mask])
```

```python
correlation = np.corrcoef(amount_array, boxes_array)
print("Correlation between Amount and Boxes Shipped:\n", correlation)
```

```python
summary = {
    'Total': np.sum(amount_array),
    'Mean': np.mean(amount_array),
    'Std Dev': np.std(amount_array),
    'Min': np.min(amount_array),
    'Max': np.max(amount_array),
}
print("Summary:\n", summary)
```
